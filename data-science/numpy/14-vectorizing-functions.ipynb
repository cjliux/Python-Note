# 向量化函数

![](https://raw.githubusercontent.com/ds19991999/githubimg/master/picgo/20180718225802.png)

sinc函数具体见：[Sinc](https://zh.wikipedia.org/wiki/Sinc%E5%87%BD%E6%95%B0)

```python
# 自定义的 sinc 函数：
import numpy as np

def sinc(x):
    if x == 0.0:
        return 1.0
    else:
        w = np.pi * x
        return np.sin(w) / w
```

```python
sinc(0.0)
```

```python
sinc(3.0)
```

```python
# 但这个函数不能作用于数组：
x = np.array([1,2,3])
sinc(x)
```

```python
# 可以使用 numpy 的 vectorize 将函数 sinc 向量化，产生一个新的函数：
vsinc = np.vectorize(sinc)
vsinc(x)
# 相当于为 x 中的每一个值调用 sinc 函数：
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-5,5,101)
plt.plot(x,vsinc(x))
```

```python
x = np.linspace(-5,5,101)
plt.plot(vsinc(x))
```

因为这样的用法涉及大量的函数调用，因此，向量化函数的**效率并不高**。
