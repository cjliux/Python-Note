# Matplotlib 基础

Matplotlib是一个类似Matlab的工具包，主要用来画图，主页地址为：[Matplotlib](https://matplotlib.org/)

````python
# 导入 matplotlib 和 numpy：
%pylab
````

## plot 二维图

```python
plot(y)
plot(x, y)
plot(x, y, format_string)
```
只给定 y 值，默认以下标为 x 轴：

````python
%matplotlib inline
x = linspace(0,2*pi,50)
plot(sin(x)) # 没有给定x,则范围为0-50
````

````python
# 给定x和y值
plot(x, sin(x)) # 给定x，则范围为0-2pi
````

````python
# 多条数据线
plot(sin(x)/x,
     x,sin(2*x))
````

````python
# 使用字符串，给定线条参数：
plot(x, sin(x), 'r-^')
````

````python
# 多线条：
plot(x,sin(x),'b-o',
    x,sin(2*x),'r-^')
````

## scatter散点图

```
scatter(x, y)
scatter(x, y, size)
scatter(x, y, size, color)
```
假设我们想画二维散点图：

````python
plot(x, sin(x), 'bo')
````

````python
# 使用 scatter 达到同样的效果
scatter(x, sin(x))
````

````python
# scatter函数与Matlab的用法相同，还可以指定它的大小，颜色等参数
x = rand(200)
y = rand(200)
size = rand(200) * 30
color = rand(200)
scatter(x, y, size, color)
# 显示颜色条
colorbar()
````

## 多图

````python
# 使用figure()命令产生新的图像：
t = linspace(0, 2*pi, 50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
````

````python
# 或者使用 subplot 在一幅图中画多幅子图：
# subplot(row, column, index)
subplot(1, 2, 1)
plot(x)
subplot(1, 2, 2)
plot(y)
````


## 向图中添加数据

````python
# 默认多次 plot 会叠加：
plot(x)
plot(y)
````

````python
# 跟Matlab类似用 hold(False)关掉，这样新图会将原图覆盖：
plot(x)
hold(False)
plot(y)
# 恢复原来设定
hold(True)
````

## 标签

````python
# 可以在 plot 中加入 label ，使用 legend 加上图例：
plot(x, label='sin')
plot(y, label='cos')
legend()
````

````python
# 或者直接在 legend中加入：
plot(x)
plot(y)
legend(['sin', 'cos'])
````

## 坐标轴，标题，网格

````python
# 可以设置坐标轴的标签和标题：
plot(x, sin(x))
xlabel('radians')
# 可以设置字体大小
ylabel('amplitude', fontsize='large')
title('Sin(x)')
````

````python
# 用 'grid()' 来显示网格：
plot(x, sin(x))
xlabel('radians')
ylabel('amplitude', fontsize='large')
title('Sin(x)')
grid()
````

## 清除、关闭图像

````python
"""
==================
A simple Fill plot
==================

This example showcases the most basic fill plot a user can do with matplotlib.
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

fig, ax = plt.subplots()

ax.fill(x, y, zorder=10)
ax.grid(True, zorder=5)
plt.show()
````

````python
"""
========================
A more complex fill demo
========================

In addition to the basic fill plot, this demo shows a few optional features:

    * Multiple curves with a single command.
    * Setting the fill color.
    * Setting the opacity (alpha value).
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500)
y1 = np.sin(x)
y2 = np.sin(3 * x)

fig, ax = plt.subplots()
ax.fill(x, y1, 'b', x, y2, 'r', alpha=0.3)
plt.show()
````
