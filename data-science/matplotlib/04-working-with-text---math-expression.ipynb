## 演示

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

在字符串中使用一对 `$$` 符号可以利用 `Tex `语法打出数学表达式，而且并不需要预先安装 `Tex`。在使用时我们通常加上` r `标记表示它是一个原始字符串`（raw string）`

````python
# plain text
plt.title('alpha > beta')

plt.show()
````

````python
# math text
plt.title(r'$\alpha > \beta$')

plt.show()
````

## 上下标`_^`

* `r'$\alpha_i > \beta_i$'`：$\alpha_i > \beta_i$
* `r'$\sum_{i=0}^\infty x_i$'`：$\sum_{i=0}^\infty x_i$
* 希腊字母和特殊符号可以用` '\ + 对应的名字'` 来显示
* ` {} `中的内容属于一个部分；要打出花括号是需要使用` \{\}`

```

```

* `Tex` 语法默认忽略空格，要打出空格使用 `'\ '`
* `\sin `默认显示为 `Roman `字体

## 例子

````python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t,s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)

plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',
         fontsize=20)

plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
````
